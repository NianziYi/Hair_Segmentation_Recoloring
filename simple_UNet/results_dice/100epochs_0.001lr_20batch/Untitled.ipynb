```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Uni\SS23\Praktikum\Hair_Segmentation_Recoloring\simple_UNet\results_dice\100epochs_0.001lr_20batch\training_measures.csv")
```

```python
plt.plot(df['Train Acc'])
plt.title('Train Accuracy', fontsize=15)
plt.xlabel('epoch', fontsize=13)
plt.ylabel('accuracy', fontsize=13)
plt.grid()
plt.savefig('Train_Acc.png')
```

```python
plt.plot(df['Train Loss'])
plt.plot(df['Valid Loss'])
plt.legend(['Train', 'Validation'], loc='upper right')
plt.title('Model Loss', fontsize=15)
plt.xlabel('epoch', fontsize=13)
plt.ylabel('loss', fontsize=13)
plt.grid()
plt.savefig('Model_Loss.png')
```
